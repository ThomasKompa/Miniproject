```python
#Creating a function to alter the outliers
def remove_outlier(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5*IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
#Session_Duration (hours)
lrduration, urduration = remove_outlier(df['Session_Duration (hours)'])
 #capping upper outliers 
df['Session_Duration (hours)'] = np.where(df['Session_Duration (hours)']
>urduration, urduration,df['Session_Duration (hours)'])
#capping lower outliers
df['Session_Duration (hours)'] = np.where(df['Session_Duration (hours)']
<lrduration, lrduration,df['Session_Duration (hours)'])

#Calories_Burned
lrBurned, urBurned = remove_outlier(df['Calories_Burned'])
df['Calories_Burned'] = np.where(df['Calories_Burned']
>urBurned, urBurned,df['Calories_Burned'])
df['Calories_Burned'] = np.where(df['Calories_Burned']
<lrBurned, lrBurned,df['Calories_Burned'])

#BMI
lrduration, urduration = remove_outlier(df['BMI'])
df['BMI'] = np.where(df['BMI']
>urduration, urduration,df['BMI'])
df['BMI'] = np.where(df['BMI']
<lrduration, lrduration,df['BMI'])


```

```python
#Numerical values

median1 = df["Age"].median()
median2 = df["Weight (kg)"].median()
median3 = df["Height (m)"].median()
median4 = df["Max_BPM"].median()
median5 = df["Avg_BPM"].median()
median6 = df["Resting_BPM"].median()
median7 = df["Session_Duration (hours)"].median()
median8 = df["Calories_Burned"].median()
median9 = df["Workout_Frequency (days/week)"].median()



df["Age"].replace(np.nan,median1,inplace = True)
df["Weight (kg)"].replace(np.nan,median2,inplace = True)
df["Height (m)"].replace(np.nan,median3,inplace = True)
df["Max_BPM"].replace(np.nan,median4,inplace = True)
df["Avg_BPM"].replace(np.nan,median5,inplace = True)
df["Resting_BPM"].replace(np.nan,median6,inplace = True)
df["Session_Duration (hours)"].replace(np.nan,median7,inplace = True)
df["Calories_Burned"].replace(np.nan,median8,inplace = True)
df["Workout_Frequency (days/week)"].replace(np.nan,median9,inplace = True)
```

```python
#Categorical Values
mode1 = df["Gender"].mode().values[0]
mode2 = df["Workout_Type"].mode().values[0]


df["Gender"] = df["Gender"].replace(np.nan,mode1)
df["Workout_Type"] = df["Workout_Type"].replace(np.nan,mode2)
```
